```python
# 10 Points

# Convert the 'Gender' categorical feature to numerical variables (Male==1, Female==0)
#After conversion, see the general info of the dataset with df.info()
#Answer

df.Gender.replace('Male',1, inplace=True)
df.Gender.replace('Female',0, inplace=True)
df.info()
```

```python
# 15 Points
# now split the data into train and test (choose test value as 0.3)

from sklearn.model_selection import train_test_split



# Answer
Xtrain, Xtest, ytrain, ytest = train_test_split(df[['Height','Weight']], 
                                              df.Gender,test_size=0.3)
```

```python
# 10 Points
# apply logistic regression model ('LogisticRegression' is imorted from sklearn for you)
from sklearn.linear_model import LogisticRegression

logreg=LogisticRegression()

# Answer
logreg.fit(Xtrain,ytrain)
```

```python
# 10 Points
# test for accuracy on the testing set ('accuracy_score' is imorted from sklearn for you)
# round the result to 2 decimals
from sklearn.metrics import accuracy_score

# Answer
round(accuracy_score(logreg.predict(Xtest), ytest),2)
```

```python
# 10 Points

# Calculate the training accuracy, round the result to 2 decimals
training_accuracy = logreg.score(Xtrain, ytrain)
round(training_accuracy,2)
```
